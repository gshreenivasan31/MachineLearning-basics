```python
import numpy as np
import math


x = np.array([1,2,3,4,5]) 
y = np.array([5,7,9,11,13])

```

```python
def gradient_descent(x,y):
    m = b = 0
    n = len(x)
    alpha = 0.0001
    it = 10000
    cost_previous = 0
    for i in range(it):
        y_pred = m*x+b
        cost = (1/n) * sum ((y-y_pred)**2)
        dm = -(2/n) * sum(x*(y-y_pred))
        db =  -(2/n)*sum(y-y_pred)
        m = m - alpha*dm
        b = b - alpha*db
        if np.allclose(cost, cost_previous, rtol=1e-20):
            break

        cost_previous = cost
        print(f"m {m}  b  {b}  iteration {i}  cost  {cost}" )
```

```python
math = np.array([92,56,88,70,80,49,65,35,66,67])
cs   = np.array([98,68,81,80,83,52,66,30,68,73])
```

```python
gradient_descent(math,cs)
```

```python
gradient_descent(x,y)
```
